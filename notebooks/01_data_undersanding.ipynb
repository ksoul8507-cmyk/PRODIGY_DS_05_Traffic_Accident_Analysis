# Data Understanding

The dataset used in this analysis is the US Accidents dataset, which contains detailed records of traffic accidents across the United States.

Each row represents a single accident, with information describing:

- Time and location of the accident
- Severity level of the accident
- Weather and environmental conditions
- Road infrastructure characteristics
- Visibility and atmospheric conditions

The dataset includes both numerical and categorical variables, making it suitable for exploratory data analysis and potential machine learning applications.

---

## Dataset Structure

The dataset contains 100,000 records and multiple features such as:

- Temporal features (Start_Time, End_Time)
- Geographic features (Start_Lat, Start_Lng, City, State)
- Weather conditions (Temperature, Humidity, Wind Speed, etc.)
- Road features (Junction, Traffic Signal, Crossing)
- Target variable: Severity
